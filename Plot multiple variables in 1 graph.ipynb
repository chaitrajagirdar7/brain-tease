```python
mp.plot(Targetfile['Age'],Targetfile['Date'] ,color='green', marker='o', linestyle='dashed',
...      linewidth=10, markersize=20,label='Age')
mp.plot(Targetfile['Weight'],Targetfile['Date'] ,color='red', marker='+', linestyle='dashed',
...      linewidth=10, markersize=20,label='Weight')
mp.plot(Targetfile['BloodPressure'],Targetfile['Date'] ,color='blue', marker='+', linestyle='dashed',
...      linewidth=10, markersize=20,label='Bp')

mp.xlabel('Age vs Weight vs bp')
mp.ylabel('Date')
mp.title('Date vs Age vs Weight vs bp')

mp.legend()
```
